```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_crypto = pd.read_csv("crypto_data.csv")
df_crypto
```

```python
df_crypto = df_crypto[df_crypto.IsTrading != False]
df_crypto = df_crypto.drop(['IsTrading'], axis=1)
df_crypto
```

```python
df_crypto = df_crypto.dropna()
df_crypto
```

```python
df_crypto = df_crypto[df_crypto.TotalCoinsMined > 0]
df_crypto
```

```python
df_crypto = df_crypto.drop(['CoinName'], axis=1)
df_crypto
```

```python
# I am dropping the unnamed: 0 column because I recieved a type econvertion error
#additionally, the column is not being used so I dropped it for that reason
df_crypto_dummy = pd.get_dummies(df_crypto, columns=['Algorithm','ProofType'])
df_crypto_dummy= df_crypto_dummy.drop(['Unnamed: 0'], axis=1)
df_crypto_dummy
```

The number columns increased significanlty in this step. It seems like, there is a column relating to algorithm and proofType for all unique values of the coin

```python
from sklearn.preprocessing import StandardScaler
crypto_scaled = pd.DataFrame(StandardScaler().fit_transform(df_crypto_dummy))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=.90)
crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate=35)
tsne_features = tsne.fit_transform(crypto_scaled)
```

```python
tsne_features.shape
```

```python
crypto_scaled['x'] = tsne_features[:,0]
crypto_scaled['y'] = tsne_features[:,1]
```

```python
plt.scatter(crypto_scaled['x'], crypto_scaled['y'])
plt.show()
```

```python
from sklearn.cluster import KMeans
crypto = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_scaled)
    crypto.append(km.inertia_)

# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "crypto": crypto}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['crypto'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Based on the elbow curve that I derived, it looks like cryptocurrencies can be clustered together. The optimal amount of clusters is 4.
